```python
#Your code here:
import numpy as np
from matplotlib import pyplot as plt
import scipy as sp
```

```python

def create_fourier_transform(signal, fs):
    "Find the spectrum of a discrete signal via fft"
    fourier = np.fft.fft(signal)
    freq = np.fft.fftfreq(signal.size, 1/fs)
    return freq, fourier
    
    
    
    
```

```python
freq = 5
tT = 1
fs = 256
time = np.linspace(0, tT, num=256)
signal = np.sin(2*5*np.pi*time)
freq, fourier = create_fourier_transform(signal, fs)
fig, (time_domain, freq_domain) = plt.subplots(2, 1)
time_domain.set_xlabel("t/s")
time_domain.set_ylabel(r'$\sin(2 \pi 5 )$')
freq_domain.set_xlabel("f/Hz")
freq_domain.set_ylabel("spectrum value")
freq_domain.set_xlim((-10, 10))
fig.tight_layout()
time_domain.plot(time, signal, '*')
freq_domain.plot(freq, fourier)
```
